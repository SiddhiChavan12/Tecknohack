```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Iris.csv')
df.describe()
```

```python
null=df.isnull().sum()
print("null values from the given dataset",null)
```

```python
df.info()
```

```python
plt.hist(df['PetalLengthCm'], bins=30, color='skyblue', edgecolor='black')
plt.ylabel('petal len')
plt.xlabel('id')
plt.title('Height of petal')
plt.show()
```

```python
plt.hist(df['PetalWidthCm'], bins=30, color='pink', edgecolor='black')
plt.xlabel('flower id')
plt.ylabel('sepal lenthr')
plt.title('Width of sepal')
```

```python
plt.hist(df['SepalLengthCm'], bins=30, color='green', edgecolor='black')
plt.xlabel('flower id')
plt.ylabel('sepal lenthr')
plt.title('Height of sepal')
```

```python
plt.hist(df['SepalWidthCm'], bins=30, color='brown', edgecolor='black')
plt.xlabel('flowerid')
plt.ylabel('sepal lenthr')
plt.title('Width of sepal')
```

```python
data = pd.Series(df['SepalLengthCm'])
data1 = pd.Series(df['SepalWidthCm'])
plt.hist([data, data1], bins=30, color=['green', 'brown'], edgecolor='black')
plt.xlabel('flower id')
plt.ylabel('length and width ')
plt.title('histogram for Sepal')
plt.legend(['length','Width'])
plt.show()
```

```python
data = pd.Series(df['PetalLengthCm'])
data1 = pd.Series(df['PetalWidthCm'])
plt.hist([data, data1], bins=30, color=['skyblue', 'pink'], edgecolor='black')
plt.xlabel('flower id')
plt.ylabel('length and width ')
plt.title('histogram for Petal')
plt.legend(['length','Width'])
plt.show()
```
